```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter (misalnya, customer risk dan developer risk)
alpha = 0.1  # customer risk
beta = 0.1    # developer risk
gamma = 0.6   # intensity failure objective

# Menghitung nilai A dan B
A = np.log(beta / (1 - alpha))
B = np.log((1 - beta) / alpha)

# Data kegagalan
failure_numbers = np.array([1, 2, 3, 4, 5, 6, 7, 8])
measures = np.array([9, 19, 43, 62, 76, 93, 102, 118])
normalized_measures = np.array([0.6, 1.26, 2.87, 4.13, 5.06, 6.20, 6.80, 7.86])

# Batas area accept dan reject
n_values = np.arange(0, 16, 0.1)
Tn_accept = (A / (1 - gamma)) - (np.log(gamma) / (1 - gamma)) * n_values
Tn_reject = (B / (1 - gamma)) - (np.log(gamma) / (1 - gamma)) * n_values

# Intersep
Tn_accept_intercept = A / (1 - gamma)
n_reject_intercept = B / np.log(gamma)

# Plot grafik
plt.figure(figsize=(10, 8))

# Plot batas area accept dan reject
plt.plot(Tn_accept, n_values, 'r--', label='Accept Boundary')
plt.plot(Tn_reject, n_values, 'g--', label='Reject Boundary')

# Plot area
plt.fill_betweenx(n_values, 0, Tn_accept, where=(Tn_accept >= 0), color='red', alpha=0.3, label='Accept')
plt.fill_betweenx(n_values, Tn_accept, Tn_reject, color='yellow', alpha=0.3, label='Continue Test')
plt.fill_betweenx(n_values, Tn_reject, max(Tn_reject), color='green', alpha=0.3, label='Reject')

# Plot observed failures
plt.plot(normalized_measures, failure_numbers, 'ko-', label='Observed Failures')

# Label, title, legend, and grid
plt.xlabel('Number of input events in normalized Usage Units')
plt.ylabel('Failure Number')
plt.title('Reliability Demonstration Chart')
plt.legend()
plt.grid(True)
plt.xlim(0, 16)
plt.ylim(0, 15)

# Annotate points
for i, (x, y) in enumerate(zip(normalized_measures, failure_numbers)):
    plt.text(x, y, str(failure_numbers[i]), fontsize=9, ha='right')

# Show plot
plt.show()
```
